```python
import numpy as np
import matplotlib.pyplot as plt
import skfuzzy as fz
U    = np.arange(-38.0,38.1,0.01) 
edn = fz.trapmf(U,[-38,-38,-14,0])
edz  = fz.trimf(U, [-14,0,14])
edp = fz.trapmf(U,[0,14,38,38])

W = np.arange(-15,15.1,0.01)
den = fz.trapmf(W, [-15,-15,-5,0])
dez = fz.trimf(W, [-5,0,5])
dep = fz.trapmf(W, [0,5,15,15])

V = np.arange(55,125.1,0.1)
ab = fz.trapmf(V,[55,55,75,90])
am  = fz.trimf(V,[75,90,105])
aa = fz.trapmf(V,[90,105,125,125])
```

```python
plt.plot(U,edn,U,edz,U,edp)
plt.title('Conjuntos difusos error')
plt.xlabel('U')
plt.ylabel('μ(e)')
```

```python
plt.plot(W,den,W,dez,W,dep,)
plt.title('Conjuntos difusos derivada del error')
plt.xlabel('W')
plt.ylabel('μ(d)')
```

```python
plt.plot(V,aa,V,am,V,ab)
plt.title('Conjuntos difusos angulo del servo')
plt.xlabel('V')
plt.ylabel('μ(a)')
```
